```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim import corpora
from gensim.models import LdaModel
from gensim.models import CoherenceModel
```

```python
from google.colab import drive

# 连接到Google Drive
drive.mount('/content/drive')

# 指定数据文件路径
data_file_path = '/content/drive/MyDrive/Colab Notebooks/cleaned_subdata.csv'  # 请替换为实际的数据文件路径
```

```python
import chardet

# 检测文件编码
with open(data_file_path, 'rb') as rawdata:
    result = chardet.detect(rawdata.read())

# 使用检测到的编码来读取文件
data = pd.read_csv(data_file_path, encoding=result['encoding'])
```

```python
# 预处理文本数据
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    words = word_tokenize(text.lower())  # 分词并转换为小写
    words = [word for word in words if word.isalpha() and word not in stop_words]  # 去除停用词和非字母字符
    return words
```

```python
# 对文本数据进行预处理
data['processed_text'] = data['review_text'].apply(preprocess_text)
```

```python
# 运行LDA-TA主题模型分析
dictionary = corpora.Dictionary(data['processed_text'])
corpus = [dictionary.doc2bow(text) for text in data['processed_text']]
num_topics = 20  # 选择主题数
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=30)
```

```python
topics = lda_model.print_topics(num_words=20)
for topic in topics:
    print(topic)
```
